## Задание 6

```python
def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
```

```python
def production_year(ddf):
    ddf['year'] = ddf.apply(lambda row: row.title[-5:-1] if RepresentsInt(row.title[-5:-1]) else 1990, axis=1)
production_year(ddf)
ddf['avgrating'] = 0
for i in range(ddf.size):
    ddf['avgrating'].iloc[i] = df[df['movieId'] == ddf['movieId'].iloc[i]]['rating'].sum()/df[df['movieId'] == ddf['movieId'].iloc[i]]['rating'].count() 
```
